### Policia

los atributos son datos referentes a la clase Policía
las funciones son las acciones que realiza la Policia clase Policía

clase Policía
3 atributos
3 funciones()

Los atributos de una Policía son
placa con nombre,
Equipo Complementario,
edad,
insignia,
uniforme Autorizado
cuales son las funciones de una Policía

Control de Identidad ()

Prevención y Custodia ()

Policía 1
Policía 2

```python
class Policia:
    # El método __init__ es el constructor que define los atributos
    def __init__(self, placa_nombre, equipo_complementario, uniforme_autorizado):
        # Atributos (Datos)
        self.placa_nombre = placa_nombre
        self.equipo_complementario = equipo_complementario
        self.uniforme_autorizado = uniforme_autorizado

    # Funciones (Acciones/Métodos)
    def control_de_identidad(self):
        print(f"[{self.placa_nombre}]: Iniciando Control de Identidad. Por favor, presente su documentación.")

    def investigacion_y_contexto(self):
        print(f"[{self.placa_nombre}]: Analizando la escena. Equipo en uso: {self.equipo_complementario}.")

    def identificacion_de_personas(self):
        print(f"[{self.placa_nombre}]: Verificando identidad en el sistema con el uniforme {self.uniforme_autorizado}.")

    def prevencion_y_custodia(self):
        print(f"[{self.placa_nombre}]: Realizando patrullaje preventivo para garantizar la seguridad.")

# --- Creación de Objetos (Instancias) ---

# Policía 1: Datos de ejemplo
policia1 = Policia("Oficial Juan Pérez - ID 4502", "Radio y Esposas", "Azul Táctico")

# Policía 2: Datos de ejemplo
policia2 = Policia("Sargento Ana Torres - ID 9910", "Kit de Primeros Auxilios", "Gala Administrativa")

# --- Ejecución de las funciones ---

print("--- ACTIVIDAD POLICÍA 1 ---")
policia1.control_de_identidad()
policia1.prevencion_y_custodia()

print("\n--- ACTIVIDAD POLICÍA 2 ---")
policia2.investigacion_y_contexto()
policia2.identificacion_de_personas()

print("fin del programa")
print("Programa creado por Daniel Nieves NL = 41 NC=1343")
```
